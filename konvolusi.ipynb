```python
import matplotlib.pyplot as plt
%matplotlib inline

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
import numpy as np

import cv2
```

```python
# Membaca citra menggunakan imread dari OpenCV
citra1 = imread(fname="cat.png")

# Menampilkan dimensi citra
print(citra1.shape)

# Menampilkan citra input dalam skala abu-abu
plt.imshow(citra1, cmap='gray')
```

```python
# Membentuk kernel untuk filter
kernel = np.array([[-1, 0, -1], 
                   [0, 4, 0], 
                   [-1, 0, -1]])

# Memproses citra menggunakan filter kernel
citraOutput = cv2.filter2D(citra1, -1, kernel)

# Menampilkan citra input dan output dalam dua plot berdampingan
fig, axes = plt.subplots(1, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(citra1, cmap='gray')
ax[0].set_title("Citra Input")

ax[1].imshow(citraOutput, cmap='gray')
ax[1].set_title("Citra Output")
```
