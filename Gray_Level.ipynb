```python
import cv2
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Membuat citra input (data.grass() adalah contoh citra yang digunakan)
img = data.grass()

# Mendapatkan dimensi citra
row, column = img.shape

# Membuat citra output dengan ukuran yang sama dengan citra input
img1 = np.zeros((row, column), dtype='uint8')

# Rentang intensitas minimum dan maksimum
min_range = 10
max_range = 60

# Proses thresholding
for i in range(row):
    for j in range(column):
        if img[i, j] > min_range and img[i, j] < max_range:
            img1[i, j] = 255
        else:
            img1[i, j] = 0
```

```python
# Menampilkan citra input dan output, serta histogramnya
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title("Citra Input")
ax[1].hist(img.ravel(), bins=256)
ax[1].set_title('Histogram Input')

ax[2].imshow(img1, cmap=plt.cm.gray)
ax[2].set_title("Citra Output")
ax[3].hist(img1.ravel(), bins=256)
ax[3].set_title('Histogram Output')

plt.show()
```
